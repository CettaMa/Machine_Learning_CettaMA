```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt

# Membaca dataset
data = pd.read_csv("/content/Salary_Data_Simple_SVR(1).csv")

# Explorasi dataset
data.head()
print()
data.info()
print()
print(data)
```

```python
# Memisahkan fitur (X) dan target (y)
X = data.iloc[:, :-1].values
y = data.iloc[:, 1].values

# Membagi data menjadi data latih dan data uji
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Membuat model regresi linear
regressor = LinearRegression()

# Melatih model regresi linear menggunakan data latih
regressor.fit(X_train, y_train)

# Memprediksi nilai gaji berdasarkan pengalaman kerja pada data uji
y_pred = regressor.predict(X_test)
```

GPT Prompt : "visualisasikan menggunakan matplotlib"

```python
# Visualisasi data latih dan model regresi
plt.scatter(X_train, y_train, color='blue', label='Data Latih')
plt.plot(X_train, regressor.predict(X_train), color='red', label='Garis Regresi')
plt.title('Regresi Linear - Pengalaman vs Gaji (Data Latih)')
plt.xlabel('Pengalaman (Tahun)')
plt.ylabel('Gaji')
plt.legend()
plt.show()

# Visualisasi data uji dan prediksi model
plt.scatter(X_test, y_test, color='blue', label='Data Uji')
plt.plot(X_test, y_pred, color='red', label='Prediksi Model')
plt.title('Regresi Linear - Pengalaman vs Gaji (Data Uji)')
plt.xlabel('Pengalaman (Tahun)')
plt.ylabel('Gaji')
plt.legend()
plt.show()
```
