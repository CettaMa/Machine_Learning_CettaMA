```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
diamonds = pd.read_csv("dataset\\diamonds.csv")
diamonds.info()
```

```python
# feature statistic
diamonds.describe()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='carat', y='price', data=diamonds, alpha=0.5)
plt.title('Scatter plot of Carat vs. Price')
plt.xlabel('Carat')
plt.ylabel('Price ($)')
plt.show()
```

```python
# Selecting numeric columns
numeric_columns = ['carat', 'depth', 'table', 'price', 'x', 'y', 'z']

# Box plot: numeric columns
plt.figure(figsize=(12, 8))
for i, col in enumerate(numeric_columns, start=1):
    plt.subplot(3, 3, i)
    sns.boxplot(x=diamonds[col], orient='v')
    plt.title(f'Box Plot of {col}')
    plt.ylabel('Values')
    plt.xlabel('')
    plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Heat map: correlation matrix
plt.figure(figsize=(10, 6))
corr = diamonds.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Heatmap of Correlation Matrix')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='cut', data=diamonds, palette='Set2')
plt.title('Bar Plot of Cut Frequency')
plt.xlabel('Cut')
plt.ylabel('Frequency')
plt.show()
```
