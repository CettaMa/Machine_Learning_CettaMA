```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
moto = pd.read_csv("dataset\\all_bikez_curated.csv")
moto_numeric = moto.select_dtypes(include=['float64', 'int64'])
moto_numeric.info()
```

```python
moto.describe()
```

```python
plt.figure(figsize=(12, 8))
for i, col in enumerate(moto_numeric.columns, start=1):
    plt.subplot(3, 4, i)
    sns.boxplot(x=moto_numeric[col])
    plt.title(f'Box Plot of {col}')
    plt.xlabel('')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Displacement (ccm)', y='Power (hp)', data=moto, alpha=0.5)
plt.title('Scatter plot of Displascment vs. Power')
plt.xlabel('displacment')
plt.ylabel('Power')
plt.show()
```

```python
# Compute the correlation matrix
correlation_matrix = moto_numeric.corr()

# Create the heatmap
plt.figure(figsize=(10, 8))  # Adjust size if needed
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Create the violin plot
plt.figure(figsize=(10, 6))  # Adjust size if needed
sns.violinplot(x="Power (hp)", data=moto)
plt.title('Violin Plot of ' + "Power (hp)")
plt.xlabel("Power (hp)")
plt.ylabel('Density')
plt.show()
```

```python
# Count occurrences of each category
category_counts = moto["Year"].value_counts()

# Create the bar plot
plt.figure(figsize=(10, 6))  # Adjust size if needed
category_counts.plot(kind='bar')
plt.title('Bar Plot of ' + "Year")
plt.xlabel("year")
plt.ylabel('Count')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(axis='y')  # Add gridlines on the y-axis
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Bore (mm)', y='Torque (Nm)', data=moto, alpha=0.5)
plt.title('Scatter plot of bore vs. Torquee')
plt.xlabel('Bore')
plt.ylabel('Torque')
plt.show()
```
