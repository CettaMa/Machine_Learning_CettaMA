```python
import pandas as pd
import numpy as np
from sklearn.cluster import AffinityPropagation
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('..\\dataset\\Mall_Customers.csv')

# Selecting features for clustering
X = data.iloc[:, [3, 4]].values  # We'll use Annual Income and Spending Score for clustering

# Feature scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Clustering using Affinity Propagation
cluster_model = AffinityPropagation(random_state=42)
clusters = cluster_model.fit_predict(X_scaled)
centroids = cluster_model.cluster_centers_

# Visualizing the clusters and centroids
plt.figure(figsize=(10, 6))

# Plot data points
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=clusters, cmap='viridis', s=50, marker='o', edgecolors='k', label='Data Points')

# Plot centroids
plt.scatter(centroids[:, 0], centroids[:, 1], marker='X', s=200, color='red', label='Cluster Centers')

plt.title('Clustering using Affinity Propagation')
plt.xlabel('Standardized Annual Income')
plt.ylabel('Standardized Spending Score')
plt.colorbar(label='Cluster')
plt.grid(True)
plt.legend()
plt.show()
```
